# GPT-2の実行

https://huggingface.co/abeja/gpt2-large-japanese

```python
from transformers import pipeline


generator = pipeline("text-generation", model="abeja/gpt2-large-japanese")
generated = generator(
    "昔々あるところに、",
    max_length=30, # 出力する文字列の長さ
    do_sample=True,
    num_return_sequences=3, # 何種類の結果を出力するか？
    top_p=0.95, # 上位{top_p}%の単語から選択する。
    top_k=50, # 上位{top_k}個の文章を保持
    temperature=0.7,
    pad_token_id=3
)
print(*generated, sep="\n")
```

# **[演習2] tiktokenを用いた入力テキストのトークンへの分割**

tiktokenはOpenAI のモデルで使用する高速 BPE トークナイザー

https://note.com/npaka/n/ncb4864df31c9

```python
pip install tiktoken
```

"Hello World!"をトークンに分割

```python
import tiktoken

# エンコード
enc = tiktoken.get_encoding("gpt2")
tokens = enc.encode("Hello World!")
print('トークン数: ', len(tokens))
print('トークン(ID): ', tokens)
print('トークン: ', enc.decode_tokens_bytes(tokens))
```

「こんにちは」をトークンに分割

```python
tokens = enc.encode("太郎はコンピュータを買った。")
print('トークン数: ', len(tokens))
print('トークン(ID): ', tokens)
def data2str(data):
    try:
        return data.decode('utf-8')
    except UnicodeError:
        return data
print('トークン: ', [data2str(data) for data in enc.decode_tokens_bytes(tokens)])
```

# **[演習3] spaCyを用いた入力テキストの解析**

形態素解析

```python
import spacy
from spacy.lang.ja.examples import sentences

nlp = spacy.load("ja_core_news_sm")
doc = nlp('太郎はコンピュータを買った。')
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

構文解析

```python
import spacy
from spacy import displacy

displacy.serve(doc, style="dep")
```

# **[演習4] spaCyとGPT-2による固有表現抽出**


```python
import spacy
from spacy.lang.ja.examples import sentences

nlp = spacy.load("ja_core_news_sm")
doc = nlp('昨日、山田君はASUSのコンピュータを買った。')
for ent in doc.ents:
    print(ent.text, ent.label_, ent.start_char, ent.end_char)
#for token in doc:
#    print(token.text, token.pos_, token.dep_)
```

```python
generator = pipeline("text-generation", model="abeja/gpt2-large-japanese")

#prompt="""以下の文から組織名を抽出してください。:
#鈴木くんは山田商事の社員だ。=>山田商事
#花子は春日総合病院に行った。=>春日総合病院
#昨日、山田君はASUSのコンピュータを買った。=>"""

prompt="""以下の文から人名を抽出してください。:
鈴木くんが歩いている。=>鈴木
花子が話す。=>花子
昨日、山田君はASUSのコンピュータを買った。=>"""

generated = generator(
    prompt,
    max_length=100, # 出力する文字列の長さ
    do_sample=True,
    num_return_sequences=1, # 何種類の結果を出力するか？
    top_p=0.95,
    top_k=50,
    temperature=0.1,
    pad_token_id=3
)
print(*generated, sep="\n")
```
